# **SALES PREDICTION USING PYTHON**

```python
import pandas as pd
```

## Getting Dataset

```python
path = '/content/drive/MyDrive/Datasets/Advertising.csv'
df = pd.read_csv(path)
df
```

## Checking, cleaning data

```python
df.info()
```

```python
df.describe()
```

## Dividing data into variables

```python
x = df.drop(columns=['Sales', 'Unnamed: 0'])
x
```

```python
y = df['Sales']
y
```

## splitting data into training and testing variables and training the data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

## Predicting the outcome

```python
y_pred = lr.predict(x_test)
y_pred
```

## Cross Checking the predicted values and actual values

```python
pd.DataFrame({'Actual': y_test, 'predicted': y_pred})
```

## Plotting in charts

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(10, 5))
sb.regplot(x = y_test, y = y_pred, scatter_kws={'color': 'blue'})
```

## Getting the distance between Predicted values and Actual values

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
